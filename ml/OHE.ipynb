```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('carprices.csv')
```

```python
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
```

```python
new_df = pd.concat([df,dummies],axis='columns')
```

```python
new_df.drop(columns=['Car Model','Mercedez Benz C class'],inplace=True)
```

```python
new_df
```

```python
# transform true to 1 and false to 0

new_df[['Audi A5','BMW X5']] = new_df[['Audi A5','BMW X5']].apply(lambda x:x*1)
```

```python
new_df
```

```python
from sklearn.linear_model import LinearRegression

reg= LinearRegression()
```

```python
x = new_df.drop(columns=['Sell Price($)'])
y = new_df['Sell Price($)']
```

```python
reg.fit(x,y)
```

```python
reg.predict([[45000,4,0,1]])
```
