```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
plt.xlabel('year')
plt.ylabel('PCI(US$)')
plt.scatter(df.year, df.per_capita_income,color='red')
```

```python
model = linear_model.LinearRegression()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']],df['per_capita_income'])
```

```python
sample = np.array([2000])
sample = sample.reshape(-1,1)
print(reg.predict(sample))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
sample = np.array(df.drop(columns=['per_capita_income']))
sample = sample.reshape(-1,1)
```

```python
prediction = reg.predict(sample)
```

```python
plt.xlabel('year')
plt.ylabel('PCI(USD)')
plt.scatter(df.year,prediction)

# plt.xlabel('year')
# plt.ylabel('PCI(US$)')
plt.scatter(df.year, df.per_capita_income,color='red')
```
