```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('challenge/train.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
m = df['Age'].mean(skipna=True)
df['Age'] = df['Age'].fillna(m)
```

```python
df.drop(columns=['PassengerId','Name','Cabin'],inplace=True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore',sparse_output=False).set_output(transform='pandas')
```

```python
encoded_ = ohe.fit_transform(df[['Embarked','Sex']])
```

```python
new_df = pd.concat([df,encoded_],axis='columns')
```

```python
new_df
```

```python
new_df.drop(columns=['Embarked','Sex','Ticket'],inplace=True)
```

```python
new_df.drop(columns=['Sex_1','Embarked_2','Embarked_3'], inplace=True)
```

```python
new_df
```

```python
new_df.rename(columns={'Sex_0':'Female','Embarked_0':'Cherbourg','Embarked_1':'Queenstown'},inplace=True)
```

```python
new_df
```

```python
new_df.drop(columns=['Survived'])
```

```python
x = new_df.drop(columns=['Survived']).values
y= new_df['Survived'].values
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x,y)
```

```python
test_df = pd.read_csv('challenge/test.csv')
```

```python
test_df
```

```python
test_df.drop(columns=['PassengerId','Name','Cabin'],inplace=True)
```

```python
le2 = LabelEncoder()
test_df['Sex'] = le2.fit_transform(test_df['Sex'])
test_df['Embarked'] = le2.fit_transform(test_df['Embarked'])
```

```python
encoded_2 = ohe.fit_transform(test_df[['Embarked','Sex']])
```

```python
test_df = pd.concat([test_df,encoded_2],axis=1)
test_df.drop(columns=['Embarked','Sex','Ticket'],inplace=True)
```

```python
test_df.drop(columns=['Sex_1','Embarked_2'], inplace=True)
```

```python
test_df.rename(columns={'Sex_0':'Female','Embarked_0':'Cherbourg','Embarked_1':'Queenstown'},inplace=True)
```

```python
test_df
```

```python
p = dt.predict(test_df)
```

```python
print(p)
```

```python
# Associate the id with the prediction
temp_df  = pd.read_csv('challenge/test.csv')
submission = pd.DataFrame({'PassengerId':temp_df.PassengerId,'Survived':p})
submission
```

```python
# write to csv
submission.to_csv('submission.csv',index=False)
```
