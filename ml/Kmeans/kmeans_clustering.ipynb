```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris= load_iris()
```

```python
iris.keys()
```

```python
iris.target
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df = pd.concat([df,pd.DataFrame(iris.target,columns=['target'])],axis=1)
```

0:'setosa',1:'versicolor',2:'virginica'

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='red',label='setosa')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',label='versicolor')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='green',label='virginica')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.legend()
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='red',label='setosa')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',label='versicolor')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green',label='virginica')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.legend()
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['petal length (cm)','petal width (cm)']])
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df
```

```python
cluster1 = df[df['cluster']==0]
cluster2 = df[df['cluster']==1]
cluster3 = df[df['cluster']==2]
```

```python
cluster1
```

```python



plt.scatter(cluster1['petal length (cm)'],cluster1['petal width (cm)'],color='green')
plt.scatter(cluster2['petal length (cm)'],cluster2['petal width (cm)'],color='red')
plt.scatter(cluster3['petal length (cm)'],cluster3['petal width (cm)'],color='blue')
plt.legend()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# First subplot - After KMeans
axes[0].scatter(cluster1['petal length (cm)'], cluster1['petal width (cm)'], color='blue', label='Cluster 1')
axes[0].scatter(cluster2['petal length (cm)'], cluster2['petal width (cm)'], color='green', label='Cluster 2')
axes[0].scatter(cluster3['petal length (cm)'], cluster3['petal width (cm)'], color='red', label='Cluster 3')
axes[0].legend()
axes[0].set_xlabel('Petal Length (cm)')
axes[0].set_ylabel('Petal Width (cm)')
axes[0].set_title('After KMeans')

# Second subplot - Before KMeans
axes[1].scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='green', label='Setosa')
axes[1].scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='red', label='Versicolor')
axes[1].scatter(df2['petal length (cm)'], df2['petal width (cm)'], color='blue', label='Virginica')
axes[1].legend()
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Petal Width (cm)')
axes[1].set_title('Before KMeans')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
# plot the elbow point with a vertical line
plt.axvline(x=3, color='r', linestyle='--')
```
