```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Titanic/challenge/train.csv')
```

```python
df
```

```python
df.drop(['PassengerId','Name','Cabin', 'SibSp', 'Parch', 'Embarked','Ticket'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean(skipna=True))
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(df[['Age','Fare','Pclass','Sex']],df['Survived'])
```

```python
validation_df = pd.read_csv('../Titanic/challenge/test.csv')
```

```python
id_ = validation_df['PassengerId']
```

```python
validation_df.isna().sum()
```
